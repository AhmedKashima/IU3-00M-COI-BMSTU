```python
def compute_haar_feature_2rect(region):
    """
    Вычисляет признак Хаара с двумя прямоугольниками по горизонтали.
    Белый слева, черный справа.
    """
    h, w = region.shape
    if w % 2 != 0:
        return 0  # нельзя поделить пополам

    mid = w // 2
    white = np.sum(region[:, :mid])
    black = np.sum(region[:, mid:])
    return white - black
```

```python
# Импорт необходимых библиотек
import cv2                      # Библиотека OpenCV — для работы с изображениями
import numpy as np              # Работа с массивами и матрицами
import matplotlib.pyplot as plt # Построение графиков и визуализация

# Загрузка изображения
image = cv2.imread('1.png')  # Чтение изображения из файла
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Преобразование в оттенки серого (градации серого)

# Координаты области интереса (Region of Interest, ROI)
x, y, w, h = 80, 90, 40, 20  # Верхний левый угол (x, y), ширина и высота
region = gray_image[y:y+h, x:x+w]  # Вырезаем нужную область из изображения

# Определение функции для генерации масок Хаара
def haar_feature(region, mode):
    h, w = region.shape  # Получаем высоту и ширину региона
    feature = np.zeros_like(region, dtype=np.int32)  # Инициализируем маску нулями

    # Режим '2h' — два горизонтальных блока: верх +1, низ -1
    if mode == '2h':
        mid = h // 2
        feature[:mid, :] = 1
        feature[mid:, :] = -1

    # Режим '2v' — два вертикальных блока: левый +1, правый -1
    elif mode == '2v':
        mid = w // 2
        feature[:, :mid] = 1
        feature[:, mid:] = -1

    # Режим '3h' — три горизонтальных блока: верх +1, середина -1, низ +1
    elif mode == '3h':
        third = h // 3
        feature[:third, :] = 1
        feature[third:2*third, :] = -1
        feature[2*third:, :] = 1

    # Режим '3v' — три вертикальных блока: левый +1, центр -1, правый +1
    elif mode == '3v':
        third = w // 3
        feature[:, :third] = 1
        feature[:, third:2*third] = -1
        feature[:, 2*third:] = 1

    # Режим '4' — четыре блока (2x2): шахматный порядок знаков
    elif mode == '4':
        hw, hh = w // 2, h // 2
        feature[:hh, :hw] = 1
        feature[:hh, hw:] = -1
        feature[hh:, :hw] = -1
        feature[hh:, hw:] = 1

    return feature  # Возвращаем маску

# Список режимов фильтрации
modes = ['2h', '2v', '3h', '3v', '4']
total_tiles = 5 * 5  # Всего 25 визуализаций: повторно применим фильтры

# Построение мозаики фильтров Хаара
plt.figure(figsize=(10, 10))  # Размер всей фигуры

for i in range(total_tiles):
    mode = modes[i % len(modes)]        # Повторно проходим по всем режимам
    mask = haar_feature(region, mode)   # Получаем маску по текущему режиму

    # Применение фильтра Хаара: поэлементное умножение
    result = region * mask

    # Нормализация результата для отображения (в диапазоне 0–255)
    result = cv2.normalize(result, None, 0, 255, cv2.NORM_MINMAX)

    # Отображение результата
    plt.subplot(5, 5, i + 1)             # Располагаем на сетке 5x5
    plt.imshow(result, cmap='gray')     # Отображаем как изображение в градациях серого
    plt.title(mode)                     # Название режима
    plt.axis('off')                     # Убираем оси

# Корректное размещение всех изображений без наложений
plt.tight_layout()
plt.show()  # Показываем финальную мозаику
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Загрузка изображения и преобразование в оттенки серого
image = cv2.imread('1.png')  # Чтение изображения из файла
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Преобразование в черно-белое (один канал)

# Загрузка каскада Хаара для обнаружения лиц
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)  # Поиск лиц на изображении

# Функция генерации Хаар-признаков
def haar_feature(region, mode):
    h, w = region.shape  # Получение размеров региона (высота и ширина)
    feature = np.zeros_like(region, dtype=np.int32)  # Создание массива признаков с нулями

    # Горизонтальный признак из двух частей
    if mode == '2h':
        mid = h // 2
        feature[:mid, :] = 1   # Верхняя половина
        feature[mid:, :] = -1  # Нижняя половина

    # Вертикальный признак из двух частей
    elif mode == '2v':
        mid = w // 2
        feature[:, :mid] = 1   # Левая половина
        feature[:, mid:] = -1  # Правая половина

    # Горизонтальный признак из трех частей
    elif mode == '3h':
        third = h // 3
        feature[:third, :] = 1
        feature[third:2*third, :] = -1
        feature[2*third:, :] = 1

    # Вертикальный признак из трех частей
    elif mode == '3v':
        third = w // 3
        feature[:, :third] = 1
        feature[:, third:2*third] = -1
        feature[:, 2*third:] = 1

    # Четырехчастный признак (2х2 блок)
    elif mode == '4':
        hh, hw = h // 2, w // 2
        feature[:hh, :hw] = 1       # Верх-лево
        feature[:hh, hw:] = -1      # Верх-право
        feature[hh:, :hw] = -1      # Низ-лево
        feature[hh:, hw:] = 1       # Низ-право

    return feature

# Список типов признаков
modes = ['2h', '2v', '3h', '3v']

# Названия ключевых областей лица
zone_names = ['Forehead', 'Left Eye', 'Right Eye', 'Nose', 'Mouth']

# Если лицо обнаружено
if len(faces) > 0:
    (x, y, w, h) = faces[0]  # Используем только первое обнаруженное лицо

    # Определение координат частей лица на основе позиции прямоугольника лица
    parts = {
        "Forehead": (x + w//4, y, w//2, h//6),                       # Лоб
        "Left Eye": (x + w//8, y + h//4, w//4, h//5),                # Левый глаз
        "Right Eye": (x + w//2, y + h//4, w//4, h//5),               # Правый глаз
        "Nose": (x + w//3, y + h//2 - h//10, w//3, h//5),            # Нос
        "Mouth": (x + w//4, y + 3*h//4, w//2, h//6),                 # Рот
    }

    # Подготовка окна отображения
    plt.figure(figsize=(14, 10))
    idx = 1  # Индекс подграфика

    # Для каждой части лица и каждого типа признака
    for zone_name, (px, py, pw, ph) in parts.items():
        region = gray[py:py+ph, px:px+pw]  # Вырезаем регион из изображения

        for mode in modes:
            haar = haar_feature(region, mode)  # Генерация Хаар-признака
            result = region.astype(np.int32) * haar  # Умножаем на изображение
            result = cv2.normalize(result, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)  # Нормализация результата

            # Создаем копию полного изображения и вставляем туда результат
            full_image = gray.copy()
            full_image[py:py+ph, px:px+pw] = result

            # Отображение результата на графике
            plt.subplot(len(parts), len(modes), idx)
            plt.imshow(full_image, cmap='gray')
            plt.title(f'{zone_name} - {mode}')
            plt.axis('off')
            idx += 1

    plt.tight_layout()
    plt.show()
else:
    print("Лицо не обнаружено.")
```

```python
def sliding_window_detection(image, window_size, step_size, threshold):
    """
    Детекция объекта методом скользящего окна с использованием простого Хаар-признака.

    Параметры:
    image -- изображение, по которому происходит поиск
    window_size -- размер окна (ширина, высота)
    step_size -- шаг скольжения окна по изображению
    threshold -- порог для классификации (если значение признака меньше порога — объект считается найденным)

    Возвращает:
    Список координат (x, y) верхнего левого угла окон, в которых был обнаружен объект
    """
    detected = []  # список для хранения координат обнаруженных объектов

    # Проход по изображению с заданным шагом
    for y in range(0, image.shape[0] - window_size[1], step_size):
        for x in range(0, image.shape[1] - window_size[0], step_size):
            # Вырезаем окно заданного размера
            window = image[y:y + window_size[1], x:x + window_size[0]]

            # Вычисляем значение Хаар-признака (двухпрямоугольного)
            f_val = compute_haar_feature_2rect(window)

            # Примитивная классификация: если значение меньше порога, считаем, что объект найден
            if f_val < threshold:
                detected.append((x, y))  # сохраняем координаты

    return detected  # возвращаем все найденные координаты
```

```python
# Применяем метод скользящего окна к изображению
# Параметры:
# window_size=(40, 20) — размер окна (ширина 40, высота 20 пикселей)
# step_size=10 — окно будет смещаться на 10 пикселей по горизонтали и вертикали
# threshold=1000 — пороговое значение для классификации
detections = sliding_window_detection(gray_image, window_size=(40, 20), step_size=10, threshold=1000)

# Создаем график для визуализации найденных регионов
fig, ax = plt.subplots()

# Отображаем изображение в градациях серого
ax.imshow(gray_image, cmap="gray")

# Проходим по всем найденным координатам и рисуем прямоугольники
for (x, y) in detections:
    # Создаем прямоугольник (рамку) вокруг обнаруженного региона
    rect = plt.Rectangle((x, y), 40, 20, edgecolor='red', facecolor='none', linewidth=2)
    ax.add_patch(rect)  # Добавляем прямоугольник на изображение

# Добавляем заголовок и убираем оси
plt.title("Результаты распознавания")
plt.axis("off")
plt.show()  # Показываем результат
```

```python
# Пример генерации графика ошибок (используются псевдо-данные, замените на реальные после тестов)

# Массив значений ложноположительных срабатываний (ошибка I рода)
false_positive = [0.1, 0.2, 0.3, 0.4]

# Массив значений ложноотрицательных срабатываний (ошибка II рода)
false_negative = [0.9, 0.8, 0.7, 0.6]

# Пороговые значения, при которых производилось измерение ошибок
thresholds = [500, 1000, 1500, 2000]

# Строим график зависимости ошибки I рода от порогового значения
plt.plot(thresholds, false_positive, label="Ошибка I рода")

# Строим график зависимости ошибки II рода от порогового значения
plt.plot(thresholds, false_negative, label="Ошибка II рода")

# Подписываем ось X — пороговое значение, при котором оценивается классификатор
plt.xlabel("Пороговое значение признака")

# Подписываем ось Y — уровень ошибки
plt.ylabel("Ошибка")

# Заголовок графика
plt.title("График ошибок")

# Добавляем легенду (обозначения линий)
plt.legend()

# Добавляем сетку для удобства восприятия
plt.grid(True)

# Показываем финальный график
plt.show()
```
